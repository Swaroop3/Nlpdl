```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

def show_dependencies(sentence):
    doc = nlp(sentence)
    print("\nSentence:")
    print(sentence)
    print("\nDependency Parse\n")
    print(f"{'Token':<12}{'Head':<12}{'Dependency'}")
    print("-" * 34)
    for token in doc:
        print(f"{token.text:<12}{token.head.text:<12}{token.dep_}")
    return doc

sentence = "I parsed this sentence correctly"
doc = show_dependencies(sentence)
displacy.serve(doc, style="dep")
```

```python
import torch
import torch.nn as nn

class BasicLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(BasicLSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        output, (hn, cn) = self.lstm(x)
        out = self.fc(output[:, -1, :])
        return out

# parameters
input_size = 10
hidden_size = 32
output_size = 2

model = BasicLSTM(input_size, hidden_size, output_size)

# random sequence input
x = torch.randn(5, 7, input_size)  # batch=5, seq_len=7

output = model(x)

print("Input shape:", x.shape)
print("Output shape:", output.shape)
print("Model output:\n", output)
```

```python
import torch
import torch.nn as nn

class BasicGRU(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(BasicGRU, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        output, hn = self.gru(x)
        out = self.fc(output[:, -1, :])
        return out

input_size = 10
hidden_size = 32
output_size = 2

model = BasicGRU(input_size, hidden_size, output_size)

x = torch.randn(4, 6, input_size)

output = model(x)

print("Input shape:", x.shape)
print("Output shape:", output.shape)
print("GRU Output:\n", output)
```
